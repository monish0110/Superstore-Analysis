```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/cleaned_superstore.csv')
```

```python
highest_sales_region = df.groupby('region')['sales'].sum().sort_values(ascending=False)
# print(highest_sales_region)
highest_sales_region.plot(kind= 'bar')
plt.ylabel("sales")
plt.title("Sales by Region")
plt.show()
```

### Insight

- West generated the highest sales ($739K).
- East was the second-best performing region ($692K).
- South generated the lowest sales ($392K).

```python
highest_profit_region = df.groupby('region')['profit'].sum().sort_values(ascending=False)
highest_profit_region.plot(kind= 'bar')
plt.ylabel("Profit")
plt.title("Profit by Region")
plt.show()
```

### Insight

- West generated the highest profit ($111K).
- East ranked second with $95K profit.
- Central had the lowest profit despite higher sales than South.

```python
highest_sales_category = df.groupby('category')['sales'].sum().sort_values(ascending=False)
highest_sales_category.plot(kind= 'bar')
plt.xticks(rotation=0)
plt.xlabel("Categories")
plt.title("Sales by Category")
plt.show()
```

### Insight

- Technology generated the highest sales ($840K).
- Furniture ranked second ($755K).
- Office Supplies generated the lowest sales ($732K).

```python
highest_profit_category = df.groupby('category')['profit'].sum().sort_values(ascending=False)
highest_profit_category.plot(kind= 'bar')
plt.xticks(rotation=0)
plt.xlabel("Categories")
plt.title("Profit by Category")
plt.show()
```

### Insight

- Technology was the most profitable category.
- Office Supplies ranked second.
- Furniture generated the lowest profit.

```python
loss_sub_category = df.groupby('sub_category')['profit'].sum().loc[lambda x: x < 0]
loss_sub_category.plot(kind= 'barh')
plt.xticks(rotation=0)
plt.ylabel("sub-categories")
plt.title("Loss by sub-Category")
plt.show()
```

### Insight

- Tables generated the highest losses.
- Bookcases and Supplies were also loss-making.
- These subcategories should be investigated for pricing and discount issues.

```python
highest_profit_products = df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(5)
highest_profit_products.index = highest_profit_products.index.str[:30] 
highest_profit_products.plot(kind= 'barh')
plt.xticks(rotation=0)
plt.xlabel("Categories")
plt.title("Sales by Category")
plt.show()
```

```python
#relationship between discount and profit

plt.scatter(df['discount'], df['profit'])
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount vs Profit')
plt.show()
```

### Insight

- Higher discounts generally lead to lower profits.
- Discounts above 30% are frequently associated with losses.

```python
#sales trend over time

sales_by_year = df.groupby('year')['sales'].sum()
print(sales_by_year)

sales_by_year.plot(figsize=(8,5), marker='o')
plt.xticks(sales_by_year.index)
plt.title('Sales Trend by Year')
plt.ylabel('Sales')
plt.show()
```

### Insight

- Sales showed an overall upward trend over time since 2024.
- The highest sales were recorded in 2026.

```python
#profit trends over time

profit_by_year = df.groupby('year')['profit'].sum()
print(profit_by_year)

profit_by_year.plot()
plt.xticks(profit_by_year.index)
plt.title("Profit Trend by Year")
plt.ylabel('Profit')
plt.show()
```

### Insight

- Profit increased over time alongside sales.
- Business profitability improved in later years.
